```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"D:\100\Day 35 - Time Series Analysis\stock_data.csv", parse_dates=True, index_col="Date")
df.head()
```

```python
df['Volume'].plot()
```

```python
df.plot(subplots=True, figsize=(4, 4))
```

```python
df_month = df.resample("M").mean()
 
fig, ax = plt.subplots(figsize=(16, 6))
 
ax.bar(df_month['2016':].index, 
       df_month.loc['2016':, "Volume"], 
       width=25, align='center')
```

```python
df.Low.diff(2).plot(figsize=(6, 6))
```

```python
window_size = 50
rolling_mean = df['Open'].rolling\
            (window_size).mean()
rolling_mean.plot()
```

```python
df['Change'] = df.Close.div(df.Close.shift())
df['Change'].plot(figsize=(10, 8), fontsize=16)
```
